```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.title("Poblacion de Argentina")
plt.xlabel("Years")
plt.ylabel("Personas en millones")
plt.grid(True)
plt.legend(['Data Series'])
# Set axis limits
plt.xlim(2006, 2018)
plt.ylim(1000, 4000)
# Add text
plt.text(2018, 4000, "Important Point")
plt.plot([2007,2010,2015,2018], [1100,1500,1600,4000], 'go-')
plt.show()
```

```python
# El grafico muestra el aumento de la Poblaion Argentina en millones de Personas mientras que pasan los Years. Escogi esta forma de representarlo para reflejar con claridad(utilizando una linea) el radical aumento del 2016 al 2018 de la Población.
```

```python
plt.title("Accioes en bolsa")
plt.xlabel("Empresas")
plt.ylabel("Acciones")
plt.grid(False)
plt.bar(['McDonalds','Burger King','Mostaza'], [3,7,2])
```

```python
# El grafico muestra la cantidad de Acciones de Empresas de hambuerguesa conocidadas. Escogi esta forma forma de representarlo para reflejar con claridad las diferentes proporciones de Acciones de cada Empresa.
```

```python
plt.plot([1, 2, 3, 4])
plt.title("Cohete")
plt.xlabel('Tiempo')
plt.ylabel('Velocidad')
plt.show()
```

```python
# Lo que tarda un cohete para salir de la atmósfera y a que velocidad lo hace. Elegiacos este grafico porque es el mas comun de usar cuando se habla de velocidad y tiempo.
```

```python
plt.title("Frecuencia de Radio")
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line = plt.plot(t, s, lw=2)
plt.ylim(-2, 2)
plt.show()
```
